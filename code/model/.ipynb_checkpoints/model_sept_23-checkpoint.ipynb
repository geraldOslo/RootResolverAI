```python
# Lets try a continous PAI:
from tensorflow.keras import backend as K
from tensorflow.keras.layers import Activation
from tensorflow.keras.layers import AveragePooling2D
from tensorflow.keras.layers import GlobalAveragePooling2D

def create_model():
    model = Sequential()
    model.add(Conv2D(32, (3, 3), dilation_rate=(2, 2), padding='same', activation='relu', input_shape=(256, 256, 1)))
    model.add(AveragePooling2D((2, 2)))
    model.add(Conv2D(64, (3, 3), dilation_rate=(2, 2), activation='relu'))
    model.add(GlobalAveragePooling2D())
    model.add(Flatten())
    
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1))  # for regression

    opt = Adam(learning_rate=0.001)
    model.compile(optimizer=opt, loss='mean_squared_error')

    return model

# Define the custom activation function
def custom_activation(x):
    return K.sigmoid(x) * 4 + 1  # scales between 1 and 5

# create the model
model = create_model()
model.add(Activation(custom_activation))

model.summary()

from sklearn.utils import compute_class_weight
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Extract filenames and labels
filenames = df['filename'].values
labels = df['prediction'].values

# Load images
images = []
for f in filenames:
   img_path = os.path.join(data_dir, f)
   img = load_img(img_path, color_mode='grayscale', target_size=(256, 256))
   img = img_to_array(img)
   images.append(img) 

# Convert to numpy array and normalize pixel values
images = np.array(images) / 255.0

# Split data with stratification
X_train, X_test, y_train, y_test = train_test_split(images, labels, stratify=labels, test_size=0.2) 

epochs = 30

history = model.fit(X_train, y_train, epochs=epochs, validation_data=(X_test, y_test))
```

```python
# Visualize output of continuos model:
predictions = model.predict(X_test)
rounded_predictions = np.round(predictions)

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(range(len(predictions)), predictions, label='Predicted Values', alpha=0.6, color='b')
plt.scatter(range(len(rounded_predictions)), rounded_predictions, label='Rounded Predicted Values', alpha=0.6, color='r')
plt.xlabel('Sample Index')
plt.ylabel('Prediction')
plt.title('Model Predictions vs Rounded Predictions')
plt.legend()
plt.show()

# Create a scatter plot for true labels vs. predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, predictions, label='Predicted Values', alpha=0.6, color='b')
plt.xlabel('True Labels')
plt.ylabel('Predicted Values')
plt.title('True vs Predicted Values')
plt.grid(True)

# Add line for perfect correlation
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], label='Perfect Correlation', linestyle='--', color='r')

plt.legend()
plt.show()
```

## U-Net architecture

```python
import os
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, concatenate
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.utils.class_weight import compute_class_weight

def unet_model(input_shape):
    inputs = Input(input_shape)
    conv1 = Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    conv1 = Conv2D(64, (3, 3), activation='relu', padding='same')(conv1)
    pool1 = model.add(AveragePooling2D((2, 2)))(conv1)
    
    # More down-convolutions and up-convolutions can be added here
    
    convN = Conv2D(64, (3, 3), activation='relu', padding='same')(pool1)
    upN = UpSampling2D(size=(2, 2))(convN)
    
    mergeN = concatenate([conv1, upN], axis=3)
    convN = Conv2D(64, (3, 3), activation='relu', padding='same')(mergeN)
    
    convLast = Conv2D(1, (1, 1), activation='sigmoid')(convN)  # Classification layer
    
    model = Model(inputs, convLast)
    model.compile(optimizer=Adam(lr=1e-4), loss='mean_squared_error')
    
    return model
```

```python
# Read CSV
df = pd.read_csv(os.path.join(data_dir, csv_file))

# Load images and labels
filenames = df['filename'].values
labels = df['prediction'].values

images = []
for f in filenames:
    img_path = os.path.join(data_dir, f)
    img = load_img(img_path, color_mode='grayscale', target_size=(256, 256))
    img = img_to_array(img)
    images.append(img)

# Normalize and reshape
images = np.array(images, dtype=np.float32) / 255.0
labels = np.array(labels)

# Split data
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, stratify=labels)
y_train = y_train - 1
y_test = y_test - 1

# Compute class weights
# Compute class weights  
class_weights_array = compute_class_weight('balanced', 
                                           classes=np.unique(y_train),
                                           y=y_train)

# Convert array to dictionary
class_weights = {i: class_weights_array[i] for i in range(len(class_weights_array))}
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create an instance of the ImageDataGenerator class
datagen = ImageDataGenerator(
    horizontal_flip=True,  # Horizontal flip
    rotation_range=5,  # Randomly rotate images in the range (degrees, 0 to 180)
    shear_range=0.2  # Randomly apply shearing transformations in a range
)

# Fit the data generator to your training data
datagen.fit(X_train)

# Training the model using data augmentation
model.fit(datagen.flow(X_train, y_train, batch_size=32),
          epochs=30,
          validation_data=(X_test, y_test),
          class_weight=class_weights)
```
