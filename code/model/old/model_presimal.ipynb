```python
# Lets try a continous PAI:
from tensorflow.keras import backend as K
from tensorflow.keras.layers import Activation

def create_model():
    model = Sequential()
    model.add(Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(256, 256, 1)))
    model.add(MaxPooling2D((2, 2)))
    model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(Flatten())
    
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1))  # for regression

    opt = Adam(learning_rate=0.001)
    model.compile(optimizer=opt, loss='mean_squared_error')

    return model

# Define the custom activation function
def custom_activation(x):
    return K.sigmoid(x) * 4 + 1  # scales between 1 and 5

# create the model
model = create_model()
model.add(Activation(custom_activation))

model.summary()

from sklearn.utils import compute_class_weight
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Extract filenames and labels
filenames = df['filename'].values
labels = df['prediction'].values

# Load images
images = []
for f in filenames:
   img_path = os.path.join(data_dir, f)
   img = load_img(img_path, color_mode='grayscale', target_size=(256, 256))
   img = img_to_array(img)
   images.append(img) 

# Convert to numpy array and normalize pixel values
images = np.array(images) / 255.0

# Split data with stratification
X_train, X_test, y_train, y_test = train_test_split(images, labels, stratify=labels, test_size=0.2) 

epochs = 30

history = model.fit(X_train, y_train, epochs=epochs, validation_data=(X_test, y_test))
```

### Binary model: Trying to predict a simpler diagnosis
If PAI > c (c being a value between 1 and 4)

```python
# Function to create the binary model
def create_binary_model():
    model = Sequential()
    model.add(Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(256, 256, 1)))
    model.add(MaxPooling2D((2, 2)))
    
    model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    
    model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
    
    model.add(Flatten())
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(2, activation='softmax'))  # Output layer with 2 nodes for binary classification

    opt = Adam(learning_rate=0.0001)
    model.compile(optimizer=opt, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    return model

# Assuming you have the binary model defined already
binary_model = create_binary_model()

# print the model summary
binary_model.summary()
```
